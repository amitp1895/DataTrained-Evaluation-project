```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(ds):
    vif=pd.DataFrame()
    vif["variables"]=ds.columns
    vif["VIF FACTOR"]=[variance_inflation_factor (ds.values,i) for i in range (ds.shape[1])]
    return (vif)

calc_vif(df)
```

### moving forward to predict Next_Tmax

```python
# splitting  target 
x=df.drop("Next_Tmax",axis=1)
y=df["Next_Tmax"]
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 


#saving best random state 
maxscor=0
maxRS=0
for i in range (1,50):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    from sklearn.linear_model import LinearRegression
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    sco=lm.score(x_train,y_train)
    if sco>maxscor:
        maxscor=sco
        maxRS=i

print("best accuracy  is ",maxscor*100, "with random state ",maxRS)

#note- code running before using PCA 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#linear regression 
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
pred=lm.predict(x_test)
lm_score=lm.score(x_train,y_train)*100
lm_score
```

```python
#Lasso 
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.000001)
ls.fit(x_train,y_train)
pred=ls.predict(x_test)
ls_score=ls.score(x_train,y_train)*100
ls_score
```

```python
#Ridge 
rd=Ridge(alpha=0.000001)
rd.fit(x_train,y_train)
pred=rd.predict(x_test)
rd_score=rd.score(x_train,y_train)*100
rd_score
```

```python
#elastic net
from sklearn.linear_model import ElasticNet
en=ElasticNet(alpha=0.000001)
en.fit(x_train,y_train)
pred=en.predict(x_test)
en_score=en.score(x_train,y_train)*100
en_score
```

```python
print ("lm",lm_score)
print ("rd",rd_score)
print ("en",en_score)
print ("ls",ls_score)
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
models=(lm,rd,en,ls)
for i in models:
    print (i , cross_val_score(i,x,y,cv=7).mean()*100)
```

```python
#usng hyper parameter 
from sklearn.model_selection import GridSearchCV
```

```python
parameters={"normalize":["true","false"],
            "fit_intercept":["true","false"],
             "max_iter":np.arange(1,50),
           "alpha":np.arange(1,5)}
```

```python
GCV=GridSearchCV(Ridge(),parameters,cv=5)
```

```python
rd.get_params().keys()
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
##trying best params 
Next_Tmax_model=Ridge(alpha=1, fit_intercept= "true", max_iter=1,normalize="true")
Next_Tmax_model.fit(x_train ,y_train)
pred=Next_Tmax_model.predict(x_test)
scoree=Next_Tmax_model.score(x_train,y_train)
print(scoree*100)
```

```python
#dumping best model 
import joblib
joblib.dump(Next_Tmax_model,"Next_Tmax_model.pkl")
```
